# Pandas Dummies 알아보기

```python
import pandas as pd
```

```python
s= pd.Series (list('abca'))
print(s)
```

```python
pd.get_dummies(s)
```

```python
import numpy as np
```

```python
s1 = ['a','b',np.nan]
print(s1)
```

```python
pd.get_dummies(s1,dummy_na = True)
```

```python
df = pd.DataFrame ({'A':['a','b','a'],'B':['b','a','c'],'C':[1,2,3]})
print(df)
```

```python
pd.get_dummies(df)
```

```python
pd.get_dummies(df,prefix = ['col1','col2'])
```
